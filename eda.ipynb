```python
# Count of placement status
plt.figure(figsize=(6, 4))
sns.countplot(x="PlacementStatus", data=df, palette="pastel")
plt.title("Placement Status Distribution")
plt.xlabel("Placement Status")
plt.ylabel("Count")
plt.show()

# Convert categorical columns to numerical for correlation analysis
df_encoded = df.copy()
df_encoded["ExtracurricularActivities"] = df_encoded["ExtracurricularActivities"].map({"Yes": 1, "No": 0})
df_encoded["PlacementTraining"] = df_encoded["PlacementTraining"].map({"Yes": 1, "No": 0})
df_encoded["PlacementStatus"] = df_encoded["PlacementStatus"].map({"Placed": 1, "NotPlaced": 0})

# Correlation matrix
correlation_matrix = df_encoded.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# Boxplot of CGPA vs Placement Status
plt.figure(figsize=(8, 5))
sns.boxplot(x="PlacementStatus", y="CGPA", data=df, palette="pastel")
plt.title("CGPA Distribution by Placement Status")
plt.show()

# Scatter plot for Aptitude Test Score vs CGPA
plt.figure(figsize=(8, 5))
sns.scatterplot(x="CGPA", y="AptitudeTestScore", hue="PlacementStatus", data=df, palette="coolwarm", alpha=0.6)
plt.title("Aptitude Test Score vs CGPA")
plt.xlabel("CGPA")
plt.ylabel("Aptitude Test Score")
plt.show()

# Count of students with and without extracurricular activities by placement status
plt.figure(figsize=(6, 4))
sns.countplot(x="ExtracurricularActivities", hue="PlacementStatus", data=df, palette="pastel")
plt.title("Effect of Extracurricular Activities on Placement")
plt.xlabel("Extracurricular Activities")
plt.ylabel("Count")
plt.show()

# Effect of placement training on placement status
plt.figure(figsize=(6, 4))
sns.countplot(x="PlacementTraining", hue="PlacementStatus", data=df, palette="pastel")
plt.title("Effect of Placement Training on Placement")
plt.xlabel("Placement Training")
plt.ylabel("Count")
plt.show()

# Distribution of Soft Skills Ratings among placed and not placed students
plt.figure(figsize=(8, 5))
sns.boxplot(x="PlacementStatus", y="SoftSkillsRating", data=df, palette="pastel")
plt.title("Soft Skills Rating vs Placement Status")
plt.show()

# Relationship between number of internships and placement status
plt.figure(figsize=(8, 5))
sns.countplot(x="Internships", hue="PlacementStatus", data=df, palette="pastel")
plt.title("Effect of Internships on Placement")
plt.xlabel("Number of Internships")
plt.ylabel("Count")
plt.show()

# Relationship between number of projects and placement status
plt.figure(figsize=(8, 5))
sns.countplot(x="Projects", hue="PlacementStatus", data=df, palette="pastel")
plt.title("Effect of Projects on Placement")
plt.xlabel("Number of Projects")
plt.ylabel("Count")
plt.show()
```
